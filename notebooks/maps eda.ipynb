```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from shapely.geometry import Point
import geopandas as gpd
import folium
from folium.plugins import MarkerCluster
from folium.plugins import FastMarkerCluster

%matplotlib inline
```

```python
zips = gpd.read_file('../data/zips.geojson')
print(zips.crs)
zips.head()
```

```python
zips.plot()
```

```python
leg_kwds = {'title':'Durham NC Zipcodes', 'loc':'upper right', 'bbox_to_anchor': (0,0), 'ncol':2}
zips.plot(column = 'ZIPCODE' , figsize = (10, 10),
          edgecolor = 'black',
          legend = True, legend_kwds = leg_kwds,
          color='lightblue')
plt.show();
```
