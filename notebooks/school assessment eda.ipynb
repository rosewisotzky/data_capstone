Results are suppressed where the number of valid test scores is less than 10 (this is denoted by a single asterisk *). In these files, suppression also occurs where any individual proficiency level is less than 1% or greater than 99% at the state and district level, or less than 5% or greater than 95% at the school level (this is denoted by two asterisks**).
