```python
with open('../data/keys.json') as fi:
    credentials = json.load(fi)
```

```python
api_key = credentials['api_key']
```

```python
params = {
    'api_key': api_key,
     'query': 'Philadelphia Cream Cheese'   
}
```

```python
response = requests.get(endpoint, params = params)
res = response.json()
```
